```python
def get_unpiped_text(art_text):
    matches = re.finditer(r'\[\[(.*?)\]\]',art_text)
    for match in matches:
        match_text = match.group(1)
        if '|' in match_text:
            clean_match_text = match_text.split('|')[1]
            match_text = match_text.replace('|','\|')
            match_text = match_text.replace('[','\[')
            match_text = match_text.replace(']','\]')
            match_text = match_text.replace('(','\(')
            match_text = match_text.replace(')','\)')
            art_text = re.sub(match_text,clean_match_text,art_text)
    return art_text

def get_clean_text(art_text):
    
    try:
        art_text = re.sub("<ref>.*?</ref>", "", art_text)
        art_text = re.sub("<ref name.*?/(ref)*?>", "", art_text)
        art_text = re.sub("<ref(.|\n)*?<\/ref>", "", art_text)
        
    except:
        pass
    
    try:
        art_text = re.sub("{{.*?}}", "", art_text)
        art_text = re.sub("{{[^}]*}}", "", art_text)
        art_text = re.sub("\([Ss]ource:.*?\)", "", art_text)
        art_text = re.sub("url=.*?}}", "", art_text)
        
        
        
    except:
        pass
    try:
        art_text = re.sub(r"{\|.*?\|}", "", art_text)
        art_text = re.sub(r"{\| class[^}]*}", "", art_text)
        art_text = re.sub("<imagemap[^<]*?</imagemap>", "", art_text)   # remove imagemap template pattern 
        art_text = re.sub(r"{\|[^}]*?}", "", art_text)
        #art_text = re.sub("\*\s\[.*\]", "", art_text)   # to remove reference starts with * []
        #art_text = re.sub("\*\[.*\]", "", art_text)
        
    except:
        pass
    
    
    try:
        art_text = re.sub("\n\|[^}]*?}}", "", art_text)
        art_text = re.sub("\n \|[^}]*?}}", "", art_text)
        art_text = re.sub("\n\|.*?\|}", "", art_text)
        art_text = re.sub("\n\|[^}]*?\|}", "", art_text)
    except:
        pass
    
    try:
        art_text = re.sub("<table.*?</table>", "", art_text)
        art_text = re.sub("<pre>.*?</pre>", "", art_text)
        art_text = re.sub("<tr>.*?</tr>", "", art_text)
        
        art_text = re.sub("<div.*?</div>", "", art_text)
        art_text = re.sub("<syntaxhighlight[^<]*</syntaxhighlight>", "", art_text)
        #art_text = re.sub("<syntaxhighlight.*?</syntaxhighlight>", "", art_text)
        
        
        art_text = re.sub("\[\[File:.*\]\]", "", art_text)
        art_text = re.sub("\[\[चित्र:.*\]\]", "", art_text)
        art_text = re.sub("\[\[బొమ్మ:.*\]\]", "", art_text)
        art_text = re.sub("\[\[Image:.*\]\]", "", art_text)
        art_text = re.sub("\[\[image:.*\]\]", "", art_text)
        art_text = re.sub("\[\[దస్త్రం:.*\]\]", "", art_text)
        art_text = re.sub("\[\[చిత్రం:.*\]\]", "", art_text)
        art_text = re.sub("\[http.*?\]", "", art_text)
        art_text = re.sub("<gallery[^}]*<\/gallery>", "", art_text)
    except:
        pass
    
    try:   
        art_text = re.sub(r"<!--.*?-->", "", art_text)
        art_text = re.sub(r"<!--[^>]*>", "", art_text)
        art_text = re.sub("<ref name.*?</ref>", "", art_text)
        art_text = re.sub("<ref name.*?>", "", art_text)
        
    except:
        pass
    
    try:
        art_text = get_unpiped_text(art_text)
    except:
        pass
        #print(art_text)
    try:
#         art_text = re.sub("\[\[", "", art_text)
#         art_text = re.sub("\]\]", "", art_text)
        art_text = re.sub("\{\{", "", art_text)
        art_text = re.sub("\}\}", "", art_text)
        art_text = re.sub("\(\)", "", art_text)
        art_text = re.sub("<small>", "", art_text)
        art_text = re.sub("</small>", "", art_text)
        art_text = re.sub("&nbsp;", "", art_text)
    except:
        pass 
    
    
    return art_text

def get_add_clean(art_text):
    try:
        art_text = re.sub("<table.*?</table>", "", art_text)
        art_text = re.sub("<pre>.*?</pre>", "", art_text)
        art_text = re.sub("<tr>.*?</tr>", "", art_text)
        art_text = re.sub("<.*?>", "", art_text)
        art_text = art_text.replace("'''", "")
        art_text = art_text.replace("<br>", "")
        art_text = art_text.replace("<br/>", "")
        art_text = art_text.replace("*", "")
        art_text = art_text.replace("<tt>", "")
        art_text = art_text.replace("</tt>", "")
        art_text = art_text.replace("<tr>", "")
        art_text = art_text.replace("</tr>", "")
        art_text = art_text.replace("<td>", "")
        art_text = art_text.replace("</td>", "")
        art_text = art_text.replace("<poem>", "")
        art_text = art_text.replace("</poem>", "")
        art_text = art_text.replace("<code>", "")
        art_text = art_text.replace("</code>", "")
        art_text = art_text.replace("</ref>", "")
        art_text = art_text.replace("<ref>", "")
        art_text = art_text.replace("\\'", "'")
        
    except:
        pass
    return art_text
```

```python
def add_article_to_df(articleID,title,lang,head_section,art_text,wiki_df):
    sectionid = 0
    sectionname = head_section
    small_line_count = 0
    
    para_id = 0
    try:
        lines = art_text.split('\n')
    except:
        print('Error For title : ',title)
    df = {}
    para_text = ''
    previous_para_text = ''
    for line in lines:
        if line.startswith('===='):
            pass
        
        elif line.startswith('==='):
            
            if (len(para_text) >200 and wiki_df.iloc[-1]['section_name'] == sectionname):
                wiki_df = wiki_df[:-1] # drop last row and add it with concatenated paragraph
                if para_id > 0:
                    para_id = para_id-1
                para_text = previous_para_text + ' ' + para_text
                df = {'articleID': articleID, 'article_name': title, 'lang': lang, 'section_id': sectionid, 'section_name':sectionname,'paragraph_id':para_id,'para_length':len(para_text) ,'para_text': para_text}
                wiki_df = wiki_df.append(df, ignore_index = True)
                previous_para_text = para_text
                
            para_text = ''
            previous_para_text  = ''
            try:
                sectionname = head_section + '_' + re.match('===(.*)===',line).group(1)
                sectionid = sectionid + 1
                para_id = 0
            except:
                pass
            
        elif line.startswith ('=='):
            if (len(para_text) > 200 and wiki_df.iloc[-1]['section_name'] == sectionname):
                
                wiki_df = wiki_df[:-1] # drop last row and add it with concatenated paragraph
                if para_id > 0:
                    para_id = para_id-1
                para_text = previous_para_text + ' ' + para_text
                para_text = re.sub("#'", ",", para_text)
                df = {'articleID': articleID, 'article_name': title, 'lang': lang, 'section_id': sectionid, 'section_name':sectionname,'paragraph_id':para_id,'para_length':len(para_text) ,'para_text': para_text}
                wiki_df = wiki_df.append(df, ignore_index = True)
                previous_para_text = para_text
                
            para_text = ''
            previous_para_text = ''
            try:
                head_section = re.match('==(.*)==',line).group(1)
                sectionid = sectionid + 1
                para_id = 0
                sectionname = head_section
                
            except:
                pass
            
        
        elif len(line) > 200:
            para_text = para_text + ' ' + line
            para_text = get_add_clean(para_text)
        else:
            small_line_count = small_line_count + 1
            
            if len(para_text) >= 400:
                para_text = re.sub("====(.*?)====", "", para_text)
                para_text = re.sub("===(.*?)===", "", para_text)
                para_text = re.sub("==(.*?)==", "", para_text)
                df = {'articleID': articleID, 'article_name': title, 'lang': lang, 'section_id': sectionid, 'section_name':sectionname,'paragraph_id':para_id,'para_length':len(para_text) ,'para_text': para_text}
                wiki_df = wiki_df.append(df, ignore_index = True)
                
                previous_para_text = para_text
                para_text = ''
                para_id = para_id + 1
    return wiki_df,small_line_count
```

```python
def get_url_count(article_text):
    url_count = 0
    url_reg = "url="
    url_reg1 = "\[\[.*?\]\]"
    url_reg2 = "\[http.*?\]"
    url_count = len(re.findall(url_reg, article_text))
    url_count = url_count + len(re.findall(url_reg1, article_text))
    url_count = url_count + len(re.findall(url_reg2, article_text))
    return url_count, len(article_text)


art_text = all_articles['hi_articles'][str(1)]
print(get_url_count(art_text))   
```

```python
wiki_stat_df = pd.DataFrame(columns=['articleID','article_name_hin','article_name_eng','article_name_tel','url_count_hin','url_count_eng','url_count_tel','article_len_hin','article_len_eng','article_len_tel'])
for i in range(len(all_articles['hi'].keys())):
    hin_url_count = 0
    hin_article_len = 0
    eng_url_count = 0
    eng_article_len = 0
    tel_url_count = 0
    tel_article_len = 0
    
    if i % 1000 == 0:
        print('processed artciles count : ',i)
    articleID = i
    try:
        hi_title = all_articles['hi'][str(i)]
        en_title = all_articles['en'][str(i)]
        te_title = all_articles['te'][str(i)]
    except:
        pass
    
    try:
        hin_art_text = all_articles['hi_articles'][str(i)]
        eng_art_text = all_articles['en_articles'][str(i)]
        tel_art_text = all_articles['te_articles'][str(i)]
    except:
        pass
    
    if pd.isnull(hin_art_text):
        pass
    else:
        hin_url_count,hin_article_len = get_url_count(hin_art_text)
        
    if pd.isnull(eng_art_text):
        pass
    else:
        eng_url_count,eng_article_len = get_url_count(eng_art_text)
    
    if pd.isnull(tel_art_text):
        pass
    else:
        tel_url_count,tel_article_len = get_url_count(tel_art_text)
        
    df = {'articleID': articleID, 'article_name_hin': hi_title, 'article_name_eng': en_title,'article_name_tel': te_title,'url_count_hin':hin_url_count,'url_count_eng':eng_url_count,'url_count_tel':tel_url_count,'article_len_hin':hin_article_len,'article_len_eng':eng_article_len,'article_len_tel':tel_article_len}
    wiki_stat_df = wiki_stat_df.append(df, ignore_index = True)       
    
```
